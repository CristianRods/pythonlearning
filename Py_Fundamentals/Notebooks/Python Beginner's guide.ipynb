## Data Types

Every value could be a data type in Python. 

A short remember of what are these datatypes available: everything is an object in python programming language. Therefore, 'datatypes' are classes and variables are instance of these classes.

Using the function 'type( )' we could get the class of the value of the variable.

There are several data types, as follows:

- Numbers = Integers, floating point numbers and complex numbers (this is a combination of number and 'j')

- Strings = Sequence of characteres inside single or double quotes "" or ''. STRINGS ARE INMUTABLE.

- List = A collection of several items that could have different data type. All enclosed by brackets '[ ]'. 

- Tuple = Is a list that can not be modified by nothing. All enclosed elements by parenthesis '( )'

- Dictionary = Is an unordered collection of items that are located using a correspondent pair of 'Key' - 'Value'. All elements are enclosed by braces '{key : value}'.

- Sets = A set is a unordered collection of unique items. All elements are stored as in the lists but enclosed by braces '{}'. This data types can not have repetitive elements, in case you assign a set equals values they give you back the set of unique elements.

```python
# Examples of Data types
def printer(element):
    ''' Put an element in the brackets after printer 
    and function will tell you the data type of this element
    '''
    print(type(element))
    
# int
print(type(7))

# Float
print(type(9.6))

# Complex number
print(type(17j))

# Strings
print(type('Hello world!'))

print(type(''' This is a multiline string. 
Where I can say hello to multiple universes'''))

#list 
my_list=['a','b',2,4,09.9,78,'bubble']

print(type(my_list))

#Tuple
my_tuple=(6,'python', ['a','b'])

print(type(my_tuple))

# Dictionary
my_dict={1:34,2:'a',34:'bubble',45:'python',3:1+3j}

print(type(my_dict))

# Set
my_set={1,3,4,5,3}
print(type(my_set))
```

### Graps the information in the data types

Retrieving Information from different Data Types
Due to some of the different data types seem before could store various kind of information
You could slice or index the content to retrieve only the information you need or change in some cases.

```python
# LIST
my_list=['a','b',2,4,09.9,78,'bubble']
# All elements 
print('my_list=',my_list[:])


# List are easily indexed or sliced using the slice operator '[]'
print(my_list[0:2])


# elements start in 0 (same in C++)
print(my_list[0])

# elements end in the last element
print(my_list[6])

# Lists are mutable: It means that you could change the value assign in any position of the list

my_list[6]='sponge'

print('New value of the sixth item=', my_list)

```

```python
# Strings

# You could slice a string in the same way that list, but elements are inmutables

string = 'Hola, Amigo'

# string[6] = 'A'
print("string[6]=", string[6])

# Note: Blank spaces are also strings ' '
print("string[5]=", string[5])

# Immutability: Next line returns an error.
string[5]='!' 
```

```python
# Tuples
# Same way than Lists. Using the operator [].
print(my_tuple)

# my_tuple[:] All elements
print("my_tuple[:] = ", my_tuple[:])

# my_tuple[2] = Second element, remembers it starts to count in 0.
print("my_tuple[2] = ", my_tuple[2])

# Notice that the list enclosed in the tuple is one element!

# Immutability: Next line returns an error.
my_tuple[2]='!' 
```

```python
# SET
# It always will be printed out ordered.
print(my_set)

# It is not possible to slice the sets and most of them are used to organize elements. 
```

```python
# DICTIONARIES

print(my_dict)

# we could ONLY access the values of the dict using the respective 'Key' of each element.
print('value of element stored in the key 34=',my_dict[34])

# Type of the element stored in key '45'
print(type(my_dict[45]))
```

### Convert Data Types

We could transform some data types into another for some variables. 


```python
# Python automatically transform some data types in order to avoid lost of information.
x = 3
y = 4.5

print('---------')

print(type(x))
print(type(y))
print(type(x+y))
print('---------')
# Using the predefined functions int(), float(), str().
# We could transform explicitly some variables

print(float(x))
print(int(y))
print('---------')
# You could cast the data type such as is doing in SQL for example. 
# Using only the built-in functions int(), float(), str().

number_str=('45')
print(type(number_str))
print('---------')
print(type(int(number_str)))
```

##  Variables 

You could create a variable in python, easily. You just need to called it and assign a determined value that could be string, float, integer, etc. or even the value of another variable mentioned before.

```python
x = 4
y = 78
z = 98

# or you could do it
x, y, z = (4,78,98)

# Used the word 'print' to print out any result, variable or element in python
print(x,y,z)
```

__Arimethic Operators:__


```python
# We assing a new value to x and y 

x = 5
y = 6

# ADD 
print('x + y =',x+y)

# Substract 
print('x - y =',x-y)

# Multiply 
print('x * y =',x*y)

# Divide
print('x / y =',x/y)

# Modulo
print('x // y =',x//y)

# Exponent
print('x ** y =',x**y)
```

__Logical and Comparison Operators:__

Comparison Operators always return a False or True

```python
x=3
y=5

# greater than
print('x > y is',x>y)

# Lower than
print('x < y is',x<y)

# Equal
print('x == y is',x==y)

# Not Equal than
print('x != y is',x!=y)

# Greater or Equal than
print('x >= y is',x>=y)

# Lower or Equal than
# Hence, it only needs one of the conditions is accomplished. 
print('x <= y is',x<=y)
```

```python
# Logical Operators 

# AND: and will result into True only if both the operands are True
x=3
y=5
# AND
print('x and y is',x==4 and y==5)

# OR : or will result into True if any of the operands is True.
print('x and y is',x==4 or y==5)

# NOT: not operator is used to invert the truth value
print('x and y is',x==4 or not y==5)
```

__Useful Functions for Lists I__

    len() returns how many elements are in a list.
    
    max() returns the greatest element of the list. How the greatest element is determined depends on what type objects are in the list. The maximum element in a list of numbers is the largest number. The maximum elements in a list of strings is element that would occur last if the list were sorted alphabetically. This works because the the max function is defined in terms of the greater than comparison operator. The max function is undefined for lists that contain elements from different, incomparable types.
    
    min() returns the smallest element in a list. min is the opposite of max, which returns the largest element in a list.
    
    sorted() returns a copy of a list in order from smallest to largest, leaving the list unchanged.

    join() is a string method that takes a list of strings as an argument, and returns a string consisting of the list elements joined by a separator string.
    
    append() a method to add an element in the end of the list

```python
print(len(my_list))
```

```python
# Other methods mentioned above. Returns an error, because those 
# only accept a same type of elements inside the list e.g. do not mix strings with numbers 

my_list=[1,2,3,45,78,995.0,98]
print(max(my_list))
print(min(my_list))
print(sorted(my_list))
```

```python
# Join is a method exclusive for Strings

new_str = "\t".join(["P", "Y", "C", "D"])
print(new_str)

new_str = "\n".join(["P", "Y", "C", "D"])
print(new_str)
```

```python
# The result of operations with methods in list embbeded in a list

a = [1, 5, 8]
b = [2, 6, 9, 10]
c = [100, 200]

print(max([len(a), len(b), len(c)]))
print(min([len(a), len(b), len(c)]))
```

```python
# Combine methods to get other results

names1 = ["Carol", "Albert", "Ben", "Donna"]
print(" & ".join(sorted(names1)))

print('---------','\n')

names2 = ["Carol", "Albert", "Ben", "Donna"]
names2.append("Eugenia")
print(sorted(names2))
```
